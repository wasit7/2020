```python
import pandas as pd
df=pd.read_csv('log1000.csv')
df.index=pd.to_datetime(df.time)
df=df.drop(['time'],axis=1)
df.index=df.index.to_period('1T')
a=df.head(300).plot(subplots=True,figsize=(18,8))
```

```python
df.index
```

```python
df[ (df.voltage>3.3) & (df.current>0.0)].head(200).plot(style='.',subplots=True,figsize=(18,8))
```

```python
df2=df[ (df.voltage>3.3) & (df.current>0.0)].groupby('cycle').agg(
    {
        'current': ['sum','max'],
        'voltage': ['min','max'],
        'temp': ['min','max']
    }
)
```

```python
a=df2.plot(style='.',subplots=True,figsize=(18,8))
```
